```python
import utils.stock_data as sd
import pandas 
import numpy

df = sd.StockData().get_stock_data()

df
```

```python
short_mavg = df.rolling(window=20, center=False).mean()
long_mavg = df.rolling(window=60, center=False).mean()

mavg_spread = short_mavg - long_mavg
```

```python
# What do we need to identify?
# The 20 and 60 day MA intersecting each other
# Identify stocks where the 20MA > 60MA for the past 20 days
# ABS(20MA-60MA) should be approaching 0 
# Find the stock where the momentum of ABS(20MA-60MA) approaching 0 is high

```

```python
last_month_data  = mavg_spread.tail(20)
cols = (last_month_data > 0).all()
undv_stocks = mavg_spread[cols[cols].index]


undv_stocks
```

```python
m_undv = undv_stocks.pct_change(10)
last_day_m = m_undv.tail(10)

cols = (last_day_m  < 0).all()
m_undv = m_undv[cols[cols].index]
m_undv
```

```python
import pandas as pd

undv_stocks = undv_stocks[cols[cols].index]

last_week_m = pd.DataFrame(m_undv.tail(1).T.reset_index().values, columns=['Ticker', 'M_Indicator'])
last_week_sma = pd.DataFrame(undv_stocks.tail(5).sum().reset_index().values, columns=['Ticker', 'SMA_Abs_Diff'])


last_week_m['Rank_M'] = last_week_m.M_Indicator.rank()
last_week_sma['Rank_SMA'] = last_week_sma.SMA_Abs_Diff.rank()

last_week = last_week_m.merge(last_week_sma, on='Ticker', how='left')
last_week['Overall_Rank'] = last_week.Rank_M + last_week.Rank_SMA

ranked_stocks = last_week.sort_values('Overall_Rank').reset_index(drop=True)
```
